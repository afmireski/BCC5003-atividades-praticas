Obter um aut√¥mato finito que reconhe√ßa a linguagem L = {w ‚àà { a, b}‚àó | w cont√©m uma quantidade √≠mpar de s√≠mbolos a e uma quantidade m√∫ltipla de 3 de s√≠mbolos b}.

```python
dfa07 = DFA(
    states={'q0', 'q1', 'q2', 'q3', 'q4', 'q5'},
    input_symbols={'a', 'b'},
    transitions={
        'q0': {'a': 'q3', 'b': 'q1'},  # Estado inicial: 'a's (par), 'b's (‚â° 0 mod 3)
        'q1': {'a': 'q5', 'b': 'q2'},  # 'a's (par), 'b's (‚â° 1 mod 3)
        'q2': {'a': 'q4', 'b': 'q0'},  # 'a's (par), 'b's (‚â° 2 mod 3)
        'q3': {'a': 'q0', 'b': 'q5'},  # 'a's (√≠mpar), 'b's (‚â° 0 mod 3) (Aceita√ß√£o)
        'q4': {'a': 'q2', 'b': 'q3'},  # 'a's (√≠mpar), 'b's (‚â° 1 mod 3)
        'q5': {'a': 'q1', 'b': 'q4'},  # 'a's (√≠mpar), 'b's (‚â° 2 mod 3)
    },
    initial_state='q0',
    final_states={'q3'}
)
```

```python
dfa07
```

```python
test_inputs = [
    "a", # A
    "abbb",  # A
    "ababab", # A
    "aaabbb", # A
    "bbba", # A
    "bababa", # A
    "b", # R
    "bbb", # R
    "abbba", # R
    "aaababb", # R
    "bbbaa", # R
    "ababa", # R
]

for input_str in test_inputs:
    if dfa07.accepts_input(input_str):
        print(f"'{input_str}' - ACEITO.")
    else:
        print(f"'{input_str}' - REJEITADO.")
```
